```python
import cv2
import json
import matplotlib.pyplot as plt
```

```python
key = '5f409b5d-90a4-4978-9aec-00554542b74f'
path = 'result/20200822_215958/'

with open('{}{}.json'.format(path, key), 'r') as outfile:
    data = outfile.read()
    
obj = json.loads(data)
img = cv2.imread('{}{}.jpg'.format(path, key))

print(data)
for tag in obj:
    bbox = tag['pos']
    img = cv2.rectangle(img, (bbox['x'], bbox['y']), (bbox['x'] + bbox['w'], bbox['y'] + bbox['h']), (0, 0, 255), 3)

plt.rcParams['figure.figsize'] = [15, 10]
plt.imshow(img)
plt.show()
```
